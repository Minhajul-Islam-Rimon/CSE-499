```python
!pip install opencv-python-headless

import torch
import torchvision.transforms as T
from torchvision.transforms.functional import to_pil_image
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import requests
import io
import os

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

IMG_SIZE = 256
SAMPLE_IMG_URL = "https://images.pexels.com/photos/1133957/pexels-photo-1133957.jpeg"
IMAGE_PATH = "sample_image.jpg"

try:
    r = requests.get(SAMPLE_IMG_URL)
    r.raise_for_status()
    with open(IMAGE_PATH, 'wb') as f:
        f.write(r.content)
    print(f"Sample image downloaded and saved to {IMAGE_PATH}")

    img = Image.open(IMAGE_PATH)
    print(f"Image loaded successfully, size: {img.size}")
except Exception as e:
    print(f"Failed to download or open image: {e}")

preprocess = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.ToTensor(),
    T.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

def tensor_to_pil(tensor):
    tensor = (tensor * 0.5) + 0.5
    tensor = torch.clamp(tensor, 0, 1)
    return to_pil_image(tensor.cpu())

def create_rectangular_mask(img_shape=(IMG_SIZE, IMG_SIZE), mask_size=(128, 128)):
    """Generates a rectangular mask with a random top-left corner."""
    height, width = img_shape
    mask_h, mask_w = mask_size

    top = np.random.randint(0, height - mask_h)
    left = np.random.randint(0, width - mask_w)

    mask = torch.zeros(img_shape, dtype=torch.float32)
    mask[top : top + mask_h, left : left + mask_w] = 1.0
    return mask

def create_free_form_mask(img_shape=(IMG_SIZE, IMG_SIZE), max_strokes=10, max_width=25, min_width=10):
    """
    Generates a free-form "brush stroke" mask using OpenCV.
    This simulates the irregular masks I mentioned.
    """
    height, width = img_shape
    mask = np.zeros((height, width), dtype=np.float32)

    for _ in range(np.random.randint(1, max_strokes + 1)):
        start_x = np.random.randint(0, width)
        start_y = np.random.randint(0, height)
        end_x = np.random.randint(0, width)
        end_y = np.random.randint(0, height)
        brush_width = np.random.randint(min_width, max_width + 1)

        cv2.line(mask, (start_x, start_y), (end_x, end_y), 1.0, brush_width)

    return torch.from_numpy(mask)

if os.path.exists(IMAGE_PATH):
    try:
        original_pil = Image.open(IMAGE_PATH).convert("RGB")
        original_tensor = preprocess(original_pil).to(device)

        rect_mask = create_rectangular_mask().to(device)
        ff_mask = create_free_form_mask().to(device)

        masked_rect_tensor = original_tensor * (1.0 - rect_mask)
        masked_ff_tensor = original_tensor * (1.0 - ff_mask)

        fig, axs = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle("Week 2: Preprocessing and Masking Summary", fontsize=20)

        axs[0, 0].set_title("1. Original Image", fontsize=14)
        axs[0, 0].imshow(tensor_to_pil(original_tensor))
        axs[0, 0].axis('off')

        axs[0, 1].set_title("2. Rectangular Mask (Hole=White)", fontsize=14)
        axs[0, 1].imshow(rect_mask.cpu(), cmap='gray')
        axs[0, 1].axis('off')

        axs[0, 2].set_title("3. Masked Image (Rectangular)", fontsize=14)
        axs[0, 2].imshow(tensor_to_pil(masked_rect_tensor))
        axs[0, 2].axis('off')

        axs[1, 0].set_title("1. Original Image", fontsize=14)
        axs[1, 0].imshow(tensor_to_pil(original_tensor))
        axs[1, 0].axis('off')

        axs[1, 1].set_title("2. Free-Form Mask (Hole=White)", fontsize=14)
        axs[1, 1].imshow(ff_mask.cpu(), cmap='gray')
        axs[1, 1].axis('off')

        axs[1, 2].set_title("3. Masked Image (Free-Form)", fontsize=14)
        axs[1, 2].imshow(tensor_to_pil(masked_ff_tensor))
        axs[1, 2].axis('off')

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

    except Exception as e:
        print(f"An error occurred during processing or plotting: {e}")
else:
    print(f"Sample image not found at {IMAGE_PATH}. Please re-run the download cell.")

```
